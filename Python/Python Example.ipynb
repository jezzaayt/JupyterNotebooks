```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Years and then Population (billions by one decimal pointing for ease of the graph

```python
years = [1950, 1955,
         1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
```

```python
pops = [2.5, 2.7, 3.0,
        3.3, 3.6, 4.0, 4.4, 4.8, 5.3,
        5.7, 6.1, 6.5, 6.9, 7.3] 
```

```python
deaths = [1.2, 1.7, 1.8, 2.2, 2.5,
          2.7, 2.9, 3, 3.1, 3.3, 3.5,
          3.8, 4.0, 4.3]
```

 # Line Chart Population

```python
lines = plt.plot(years, pops )
plt.grid(True)

plt.setp(lines, color=(255/255, 50/205,155/255), marker="o")

plt.ylabel("Population in Billions")
plt.xlabel("Population Growth in Years")
plt.title("Population Growth")
plt.show()
```

 # Line Chart Deaths

```python
lines2 = plt.plot(years, deaths )
plt.grid(True)

plt.setp(lines, color=(255/255, 50/205,155/255), marker="o")

plt.ylabel("Deaths in Billions")
plt.xlabel("Deaths Growth in Years")
plt.title("Death Rate")
plt.show()
```

 # Pie Chart


```python

labels = "Python", "C#", "Perl", "Ruby", "Java", "C++"
sizes = [33, 52, 12, 17, 62, 48]
separated = (.2,0,0,0,0,0.1)

plt.pie(sizes, labels = labels, autopct="%1.1f%%", explode=separated)
plt.axis("equal")
plt.show()
```

# using pandas library


```python
raw_data = {"names": ["Nick", "Jeremy", "John" ,"Panda","Ivan" ],
        "jan_ir": [143, 122, 101, 108, 365],
        "feb_ir": [122, 132, 144, 98, 62],
        "march_ir": [65,88,12,32,65]}
```

```python
df = pd.DataFrame(raw_data, columns=["names", "jan_ir", "feb_ir", "march_ir"])
df['total_ir'] = df["jan_ir"] + df["feb_ir"] + df["march_ir"]
```

```python
print(df)
```

```python
color = [(1,.4,.4), (.2,.6, 1), (.5, .3, 1), (.3, 1, .5), (.7, .7, .2)]

plt.pie(df["total_ir"],
        labels=df["names"],
        colors=color,
        autopct="%1.1f%%")

plt.axis("equal")
plt.show()
```

# function

```python
col_count = 3 # manually spacing
bar_width = .1
korea_scores = (554, 536, 538)
canada_scores = (518,523, 525)
china_scores = (613, 570, 580)
france_scores = (495, 505, 499)


k1 = plt.bar(index, korea_scores, bar_width,  label="Korea", color=(.3,.3,.7))
c1 = plt.bar(index + 0.1, canada_scores, bar_width, label="Canada", color=(1,.5,0))
c2 = plt.bar(index + 0.2, china_scores, bar_width, label="China", color=(0,1,1))
f1 = plt.bar(index + 0.3, france_scores, bar_width, label="France", color=(1,1,0))


def CreateLabels(data):
    for item in data:
        height = item.get_height()
        plt.text = (item.get_x() + item.get_width() / 2., height*1.05,
                    '%d' % int(height))
               #    ha='center', va='bottom')

CreateLabels(k1)
CreateLabels(c1)
CreateLabels(c2)
CreateLabels(f1)

plt.ylabel("Mean score in PISA 2012")
plt.xlabel("Subjects")
plt.title("Test Scores by Country")


plt.xticks(index + .3 / 2, ("Maths", "Reading", "Science"))
plt.grid(True)
plt.legend(frameon=False, bbox_to_anchor=(1,1), loc=2)




integer = 365
plt.show()
```
