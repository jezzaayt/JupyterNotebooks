```r
library(tm)
library(wordcloud)
library(stringr)
```

# Reading a text file and creating wordcloud from it
# Removing stop words and negative words

```r
readLines("textminingtest.txt")
str(readLines("textminingtest.txt"))
```

```r
text <- paste(readLines("textminingtest.txt"), collapse=" ") # collapses all the elements of the chr vector
positivewords <- scan("positive-words.txt", what="character", comment.char=";")
negativewords <- scan("negative-words.txt", what="character", comment.char=";")
```

```r
#gsub takes a pattern for regex
# \\W replaces space as space
```

```r
text2 <- gsub(pattern="\\W", replace=" ", text) 
```

```r
#\\d looks for digits 0-9
```

```r
text3 <- gsub(pattern="\\d", replace=" ", text2)
text3
```

```r
#puts all to lower case
text3 <- tolower(text3)
text3
```

```r
#cleaning the data
#removing the stop words
text3<- removeWords(text3, stopwords())
```

```r
# \\b string starts with the next letter e.g. 'c' e.g. crystal, crime
text4 <- gsub(pattern="\\b[A-z]\\b{1}", replace =" ", text3) #deletes all single letters
text4
```

```r
#stripping the white space
text4 <- stripWhitespace(text4) 
text4
```

# sentiment analysis / wordcloud


```r
#splits string into individual words
# can specific words you want to look for
wordbag <- str_split(text4, pattern="\\s+")
```

```r
class(wordbag)
wordbag <- unlist(wordbag)
```

```r

#positive words and negative words lists
sum(!is.na(match(wordbag, positivewords))) # sum of the positive terms

```

```r
sum(!is.na(match(wordbag, negativewords))) # sum of the negative terms
```

# overall sentiment score of positive - negative words

```r
score <- sum(!is.na(match(wordbag, positivewords))) - sum(!is.na(match(wordbag, negativewords)))
score
mean(score)
```

```r
wordcloud(wordbag)
```

```r
wordcloud(wordbag, min.freq=2, random.order = FALSE, scale=c(3,1), color=rainbow(5))
```
